```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
```

```python
np.random.seed(42)
X = 2 * np.random.rand(100, 1)
y = 4 + 3 * X + np.random.randn(100, 1)
```

```python
plt.figure(figsize=(8, 6))
plt.scatter(X, y, alpha=0.6, color='blue')
plt.xlabel('X')
plt.ylabel('y')
plt.title('Original Data')
plt.grid(True, alpha=0.3)
plt.show()
```

```python
model = LinearRegression()
model.fit(X, y)
```

```python
y_pred = model.predict(X)
```

```python
print(f"Intercept: {model.intercept_[0]:.4f}")
print(f"Slope: {model.coef_[0][0]:.4f}")
```

```python
mse = mean_squared_error(y, y_pred)
r2 = r2_score(y, y_pred)
```

```python
mse, r2
```

```python
plt.figure(figsize=(8, 6))
plt.scatter(X, y, alpha=0.6, color='blue', label='Data')
plt.plot(X, y_pred, color='red', linewidth=2, label='Prediction')
plt.xlabel('X')
plt.ylabel('y')
plt.title(f'Data with Linear Regression\nMSE: {mse:.4f} | R²: {r2:.4f}')
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()
```
